```python
from transformers import AutoTokenizer
```

```python
# 加载模型
model_path = r"D:\apply\python\project\pipeline\model\roberta-base-finetuned-dianping-chinese"
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
# 定义句子
sen = "弱小的我也有大梦想!"

# 对句子进行分词并转换成一个序列
tokens = tokenizer.encode(sen)
print(tokens)

# 将一个序列再转换成句子
# str_sen = tokenizer.decode(tokens)
```

```python
# 填充一个句子的序列
fill = tokenizer.encode(sen,padding="max_length",max_length=15)
print(fill)
# 截断一个句子的序列
# truncation = tokenizer.encode(sen,max_length=5,truncation=True)
# print(truncation)
```

```python
# 识别句子中哪些是用户填充的词，哪些是句子本身的词
inputs = tokenizer.encode_plus(sen,padding="max_length",max_length=15)
inputs
```

```python
# 总结
sens = ["弱小的我也有大梦想!",
        "有梦想谁都了不起",
        "追逐梦想的心，比梦想本身，更可贵"]
res = tokenizer(sens)
res
```
